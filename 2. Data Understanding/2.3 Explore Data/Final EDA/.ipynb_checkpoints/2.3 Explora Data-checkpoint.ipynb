```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
```

```python
# Combine Test and Train Datasets
train_df = pd.read_csv('Final_Train.csv')
test_df = pd.read_csv('Final_Test.csv')
df = train_df.append(test_df, ignore_index=True)
```

```python
# Assuming your data is loaded into a DataFrame named 'df'
# Replace 'path/to/your/data.csv' with the actual path to your CSV file if needed
# df = pd.read_csv('path/to/your/data.csv')

# Columns of interest
X_vars = ['Runtime', 'Genre Cluster', 'G', 'NC-17', 'NR', 'PG', 'PG-13', 'R', 'Holiday', 'Adj Merged Budget', 'Has Star Score', 'Has Director Score', 'Has Production Company Score', 'Has Domestic Distributor Score', 'Unweighted Star Score', 'Simple Weight Star Score', 'Log Weight Star Score', 'Exponential Weight Star Score', 'Total Director Score', 'Avg Director Score', 'Total Production Company Score', 'Avg Production Company Score', 'Domestic Distributor Score', 'Season_ASO_4', 'Season_FMA_2', 'Season_MJJ_3', 'Season_NDJ_1']
y = ["Adj Merged Revenue"]

# Create a folder to save the plots
output_folder = 'Plots'

# Basic statistics of numerical features
summary_statistics = df[X_vars + y].describe()
summary_statistics.to_csv(os.path.join(output_folder, 'summary_statistics.csv'))

# Pairplot to visualize relationships between numerical features
numerical_columns = df[X_vars + y].select_dtypes(include=['float64', 'int64']).columns
pairplot = sns.pairplot(df[numerical_columns])
pairplot.savefig(os.path.join(output_folder, 'pairplot.png'))
plt.close()

# Correlation heatmap to identify correlations between features
correlation_matrix = df[X_vars + y].corr()
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
heatmap.figure.savefig(os.path.join(output_folder, 'correlation_heatmap.png'))
plt.close()

# Boxplot to visualize the distribution of the target variable
boxplot = sns.boxplot(x=df['Adj Merged Revenue'])
boxplot.figure.savefig(os.path.join(output_folder, 'boxplot.png'))
plt.close()

# Violin plots for categorical variables
for var in ['Genre Cluster', 'Holiday']:
    violinplot = sns.violinplot(x=var, y='Adj Merged Revenue', data=df)
    violinplot.figure.savefig(os.path.join(output_folder, f'violinplot_{var}.png'))
    plt.close()
```

```python
# Adjust plot aesthetics for better readability
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, horizontalalignment='right')
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0)

# Save the heatmap to a file
heatmap.figure.savefig(os.path.join(output_folder, 'correlation_heatmap.png'))

# Display the plot
plt.show()
```

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming your data is loaded into a DataFrame named 'df'
# Replace 'path/to/your/data.csv' with the actual path to your CSV file if needed
# df = pd.read_csv('path/to/your/data.csv')

# Columns of interest
X_vars = ['Runtime', 'Genre Cluster', 'G', 'NC-17', 'NR', 'PG', 'PG-13', 'R', 'Holiday', 'Adj Merged Budget', 'Has Star Score', 'Has Director Score', 'Has Production Company Score', 'Has Domestic Distributor Score', 'Unweighted Star Score_normalized', 'Simple Weight Star Score_normalized', 'Log Weight Star Score_normalized', 'Exponential Weight Star Score_normalized', 'Total Director Score_normalized', 'Avg Director Score_normalized', 'Total Production Company Score_normalized', 'Avg Production Company Score_normalized', 'Domestic Distributor Score_normalized', 'Season_ASO_4', 'Season_FMA_2', 'Season_MJJ_3', 'Season_NDJ_1']
y = "Adj Merged Revenue"

# Define revenue bins
bin_edges = [0, 1000000, 5000000, np.inf]
bin_labels = ['Low Revenue', 'Medium Revenue', 'High Revenue']

# Create a new column for revenue categories
df['Revenue_Category'] = pd.cut(df[y], bins=bin_edges, labels=bin_labels, right=False)

# Loop through each revenue category
for category in bin_labels:
    category_df = df[df['Revenue_Category'] == category]

    # Pairplot to visualize relationships between numerical features
    numerical_columns = category_df[X_vars + [y]].select_dtypes(include=['float64', 'int64']).columns
    pairplot = sns.pairplot(category_df[numerical_columns])
    pairplot.savefig(os.path.join(output_folder, f'pairplot_{category.lower().replace(" ", "_")}.png'))
    plt.close()

    # Boxplot for each numerical feature
    plt.figure(figsize=(18, 12))
    for i, column in enumerate(numerical_columns, 1):
        plt.subplot(4, 7, i)
        sns.boxplot(x=category_df[column])
        plt.title(f'Boxplot of {column} - {category}')
        plt.xlabel('')
        plt.ylabel('')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f'boxplots_{category.lower().replace(" ", "_")}.png'))
    plt.close()

    # Violin plots for selected categorical variables
    categorical_variables = ['Genre Cluster', 'Holiday']
    plt.figure(figsize=(15, 8))
    for i, var in enumerate(categorical_variables, 1):
        plt.subplot(1, 2, i)
        sns.violinplot(x=var, y=y, data=category_df)
        plt.title(f'Distribution of {y} by {var} - {category}')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f'violinplots_{category.lower().replace(" ", "_")}.png'))
    plt.close()
    
    # Heatmap for correlation matrix
    correlation_matrix = category_df[X_vars + [y]].corr()

    # Set up the matplotlib figure
    plt.figure(figsize=(24, 20))  # Adjust the figure size here

    # Customize the heatmap
    sns.set(font_scale=1.2)  # Adjust the font scale
    sns.set_style("whitegrid")

    # Set the fontsize for annotations
    annot_kws = {"fontsize": 12}  # Adjust the fontsize for larger squares

    # Plot the heatmap
    heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, square=True, cbar_kws={"shrink": 0.75}, annot_kws=annot_kws)

    # Adjust plot aesthetics for better readability
    heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, horizontalalignment='right')
    heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0)

    # Save the heatmap to a file
    heatmap.figure.savefig(os.path.join(output_folder, f'correlation_heatmap_{category.lower().replace(" ", "_")}.png'))
    plt.close()  # Close the current figure to avoid overlapping with subsequent figures
```

```python
# Create a histogram of the log-transformed 'Adj Merged Revenue'
output_folder = 'Plots'

plt.figure(figsize=(8, 6))
sns.histplot(np.log(df['Adj Merged Revenue']), bins=50, kde=True)  # Applying log1p to handle potential zero values
plt.title('Distribution of Log-transformed Adj Merged Revenue')
plt.xlabel('Natural Log-transformed Adj Merged Revenue')
plt.savefig(os.path.join(output_folder, 'natural_log_revenue_hist.png'))
plt.close()
```
